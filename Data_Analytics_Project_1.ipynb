```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.colors as mcolors
import numpy as np
import statistics as stats
from datetime import date
from textwrap import wrap
import calendar
import random
```

```python
# Считаем данные из файлов, запишем в датафреймы, переименуем некоторые столбцы
courses = pd.read_csv('Data/courses.csv', index_col=0)
courses.rename(columns={'id': 'course_id',
                        'title': 'course_title',
                        'field': 'course_field'}, inplace=True) 

course_contents = pd.read_csv('Data/course_contents.csv')

progresses = pd.read_csv('Data/progresses.csv')
progresses.rename(columns={'id': 'progress_id'}, inplace=True) 

progress_phases = pd.read_csv('Data/progress_phases.csv', 
                              parse_dates=[4, 5],
                              # Преобразуем даты в формат без UTC
                              date_parser=lambda x: pd.to_datetime(x).tz_localize(None))

students = pd.read_csv('Data/students.csv', 
                       index_col=0,
                       parse_dates=[3],
                       date_parser=lambda x: pd.to_datetime(x, errors='coerce'))
students.rename(columns={'id':'student_id'}, inplace=True)
```

```python
# Соединяем последовательно все датасеты в рабочий датасет
main = progress_phases\
.merge(progresses, on='progress_id')\
.merge(courses, on='course_id')\
.merge(course_contents, on=['course_id', 'module_number', 'lesson_number'])
```

#### Общее количество курсов в датасете

```python
len(main.course_id.unique()) # Выбирем из датасета уникальные id курсов и посчитаем их количество
```

#### Количество модулей на каждом курсе

```python
# Сгруппируем датафрейм по курсам, выберем только уникальные номера модулей для каждого курса и посчитаем их количество
main.groupby('course_title')['module_number'].unique().agg(len)
```

#### Количество уроков в каждом модуле на каждом курсе

```python
# Сгруппируем по курсам и модулям, выберем уникальные номера уроков 
# для каждого модуля каждого курса и посчитаем их количество

number_of_lessons = main.groupby(['course_title', 'module_title'])['lesson_number'].unique().agg(len)
number_of_lessons
```

#### Медианное количество уроков в модуле на каждом курсе

```python
# Уберем группировку по модулям и посчитаем для каждого курса медиану
number_of_lessons.unstack(level=0).agg('median')
```

#### Количество учеников на каждом курсе

```python
# Сгруппируем датафрейм по курсам, выберем уникальные student_id для каждого курса, посчитаем их количество
# (Группировка по сфере курса нужна, чтобы сохранить значения сферы для дальнейшего построения графика) 
students_on_the_course = main.groupby(['course_title', 'course_field'])['student_id']\
                                .unique()\
                                .agg(len)\
                                .reset_index(name='number_of_students')

students_on_the_course
```

#### Минимальный, максимальный, средний, медианный возраст студентов

```python
# Функция, преобразовывающая дату рождения в возраст
def BirthDateToAge(birth_date): 
    today = date.today() 
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day)) 
    return age 
```

```python
# Выберем нужные столбцы из основного датафрейма и склеим его с датафремом students
students_ages = main[['course_title', 'student_id']].drop_duplicates().merge(students, on='student_id')
# Добавим в полученный датафрейм столбец с возрастом студента, применив написанную функцию
students_ages['age'] = students_ages['birthday'].apply(BirthDateToAge)
# Профильтруем датафрейм, отсеив значения, которые являются не совсем корректными 
students_ages_clear = students_ages[(students_ages['age'] >= 10) & (students_ages['age'] <= 80)]
students_ages_clear.head(10)
```

```python
# Найдем в столбце с возрастом минимум, максимум, среднее и медиану
students_ages_clear['age'].agg(['min', 'max', 'mean', 'median'])
```

#### Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

```python
# Сгруппируем датафрейм по курсу и посчитаем те же показатели для каждого курса
students_ages_clear.groupby('course_title')['age'].agg(['min', 'max', 'mean', 'median'])
```

Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

```python
# Для отображения сферы курса на графике, добавим в датафрейм столбец с цветом
def field_color(x):
    if x == 'Design':
        return '#D81159'
    elif x == 'Marketing':
        return '#EAC4D5'
    elif x == 'Development':
        return '#D6EADF'
    elif x == 'Business':
        return '#218380'
    
students_on_the_course['graph_color'] = students_on_the_course.course_field.apply(field_color)
```

```python
# Отсортируем курсы по количеству студентов
students_on_the_course.sort_values(by='number_of_students', inplace=True)
# Посчитаем медианное значение количества студентов
my_median = np.median(students_on_the_course.number_of_students)
```

```python
students_on_the_course.drop_duplicates(subset='course_field')
```

```python
# Создание графика
fig, ax = plt.subplots(1, 1, figsize=(16, 10))

# Отрисовка графика
ax.barh('course_title', 'number_of_students',\
            data=students_on_the_course,\
            color=students_on_the_course.graph_color)
# Добавим линию медианы
ax.axvline(my_median, color='#2D2D2A', alpha=0.5, label=f'Median = {my_median}', ls='--')
# Оформление графика
ax.set_title('Количество студентов на курсе, чел.', fontsize=18)
ax.set_frame_on(False)
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')
# Добавим подписи с количеством студентов на график
for i in ax.patches:
    ax.text(x=20, 
            y=i.get_y() + 0.25, 
            s=str(i.get_width()),
            fontsize=11, 
            color='#2D2D2A')
    
# Сформируем легенду
patches = []
for index, row in students_on_the_course.drop_duplicates(subset='course_field').iterrows():
    patches.append(mpatches.Patch(color=row.graph_color, label=row.course_field))
my_handles, _ = ax.get_legend_handles_labels()
plt.legend(handles=patches + my_handles, fontsize=12, framealpha=0.1)
plt.tick_params(labelsize=14)

plt.show()
```

1. **(p)** Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля (разность между временем начала и окончания выполнения домашней работы) для каждого курса. Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы и временем окончания выполнения последней домашней работы в модуле

```python
# Выберем из основного датафрейма записи о выполненных домашних работах с корректными датами начала и завершения,
# отсортируем их по курсам, модулям и урокам
correct_homework = main[(main['is_homework'] == True) & (main['status'] == 'done') \
                      & (main['finish_date'] > main['start_date'])]\
                   .sort_values(by=['course_title', 'module_number', 'lesson_number'])

# Отбросим ненужные столбцы
correct_homework = correct_homework[['course_title', 'module_title', 'module_number', \
                                     'lesson_number', 'student_id', 'start_date', 'finish_date']]
correct_homework.head()
```

```python
# Сгруппируем полученный датафрейм по курсам, модулям и студентам
correct_homework_grouped = correct_homework.groupby(['course_title', 'module_number', 'module_title', 'student_id'])
# Запишем ключи - названия курсов
groups_keys = correct_homework_grouped.groups.keys()
```

```python
# Создадим новый датафрейм
time_to_complete = pd.DataFrame()

for group in groups_keys:
    # Выберем одну группу
    current_group = correct_homework_grouped.get_group(group)
    # Найдем для выбранной группы дату начала первого дз и завершения последнего
    start  = current_group["start_date"].min()
    finish = current_group["finish_date"].max()
    # Запишем промежуток времени между полученными датами
    time = pd.Timedelta(finish - start)
    # Создадим отдельный датафрейм для этой группы
    current_df = pd.DataFrame({'course_title': [group[0]], 
                                'module_number': [group[1]],
                                'module_title': [group[2]],
                                'lead_time': [time]})
    # Добавим данные в общий датафрейм
    time_to_complete = pd.concat([time_to_complete, current_df], ignore_index=True)
```

```python
time_to_complete.head()
```

```python
# Добавим в датафрейм столбец с временем прохождения модуля в формате int64 для более простого вычисления
time_to_complete['lead_time_int64'] = time_to_complete['lead_time'].values.astype('int64')

# Посчитаем для каждого курса минимальное, максимальное, среднее и медианное время прохождения каждого модуля
# и преобразуем обратно в формат временного промежутка для более легкого восприятия
time_to_complete = time_to_complete\
                   .groupby(['course_title', 'module_number', 'module_title'])['lead_time_int64']\
                   .agg(['min', 'max', 'mean', 'median'])\
                   .apply(lambda x: pd.to_timedelta(x))

time_to_complete.head(9)
```

```python
# Функция установки интервала для yticks
def SetYTicks(yticks):
    if (yticks % 2 != 0):
        yticks += 1
    if (yticks >= 40):
        return 4
    elif (yticks >= 20):
        return 2
    else: return 1
```

```python
# Создание графика
fig, ax = plt.subplots(nrows=5, ncols=3, figsize=(15,50))
fig.suptitle('Медианное время выполнения домашнего задания по модулям', y=1, fontsize=16)    

for ax, course in zip(np.ravel(ax), time_to_complete_grouped.groups.keys()):
    # Выберем отдельную группу
    current_group = time_to_complete_grouped.get_group(course)
    # Преобразуем медианное время прохождения в количество дней
    median_time = current_group['median'] / pd.Timedelta('1 day')
    # Отрисовка графика для данной группы
    ax.plot(current_group['module_title'], median_time, color='#D81159')
    # Оформление графика (не будем подписывать xlabel, так как это нагромождает график еще больше)
    ax.set_title(course)
    ax.set_ylabel('Время выполнения, дней')
    ax.grid(alpha=0.5)
    # Оформление xticks
    ax.set_xticks(current_group['module_title'])
    ax.set_xticklabels(current_group['module_title'], rotation='vertical', fontsize=10) 
    # Оформление yticks
    max_ytick = round(median_time.max())
    y_range = range(0, max_ytick + 1, SetYTicks(max_ytick))
    ax.set_yticks(y_range)
    ax.set_yticklabels(y_range, fontsize=9, fontweight='roman')

plt.tight_layout()
plt.show()
```

```python
seasonality = season_time_to_complete.copy()
# Возьмем из даты завершения задания только месяц
seasonality['finish_month'] = seasonality['finish_date'].apply(lambda x: x.month)

# Добавим в датафрейм столбец с временем прохождения в формате int64 для более простого вычисления
seasonality['lead_time_int64'] = seasonality['lead_time'].values.astype('int64')

# Сгруппируем по курсам и месяцу завершения работы, посчитаем медианное время прохождения для каждого курса
# и преобразуем обратно в формат временного промежутка для более легкого восприятия,
# сбросим группировку по месяцу завершения
seasonality = seasonality.groupby(['course_title', 'finish_month'])['lead_time_int64']\
.agg('median')\
.apply(lambda x: pd.to_timedelta(x))\
.reset_index(level=1, name='median_lead_time')

# Изменим вид для месяца завершения для более красивого отображения
seasonality['finish_month'] = seasonality['finish_month'].apply(lambda x: calendar.month_abbr[x])
```

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# Найдем для каждого модуля в каждом курсе номер последнего урока с домашним заданием
last_homework_lesson = course_contents[course_contents['is_homework'] == True]\
.groupby(['course_id', 'module_number'])['lesson_number']\
.agg(last_homework_lesson = 'max')
last_homework_lesson.head()
```

```python
# Выберем из основного датафрейма только домашние задания и отсортируем из по курсам, модулям и студентам
conversion_last_homework = main[main['is_homework'] == True]\
                           .sort_values(by=['course_title', 'module_number', 'student_id'])
# Добавим номер последнего урока с домашним заданием для каждого модуля каждого курса
conversion_last_homework = conversion_last_homework.merge(last_homework_lesson, 
                                                          how='left', 
                                                          on=['course_id', 'module_number'])
# Отбросим ненужные столбцы
conversion_last_homework = conversion_last_homework[['course_field', 'course_title', 'module_title', 'module_number',
                         'lesson_number', 'last_homework_lesson', 'student_id', 'status']]
conversion_last_homework
```

```python
# Сгруппируем и найдем индексы с последним уроком с домашним заданием
last_lesson = conversion_last_homework.groupby(['course_title', 'student_id', 'module_number'])['lesson_number'].idxmax()
# Выберем по этим индексам строки из датафрейма
conversion = conversion_last_homework.loc[last_lesson]
conversion.head()
```

```python
# Добавим столбец, для которого 1 - первое дз начато, 0 - не начато 
# (если строка попала в данный датафрейм, то первое дз точно начато)
conversion['started'] = 1
# Добавим столбец, для которого 1 - последнее дз завершено, 0 - не завершено
# (найдем те уроки, которые содержат последнее дз в модуле и проверим факт их завершения)
conversion['finished'] = conversion\
.apply(lambda x: np.where(((x['lesson_number'] == x['last_homework_lesson']) & (x['status'] == 'done')), 1, 0), axis=1)
```

```python
# Сгруппируем по курсам, модулям и урокам и просуммируем все начатые первые дз и завершенные последние
conversion_started_finished = conversion\
    .groupby(['course_field', 'course_title', 'module_number', 'module_title'])[['started', 'finished']]\
    .apply(sum)\
    .reset_index()

conversion_started_finished.head()
```

```python
# Сгруппируем по курсам
conversion_started_finished_grouped = conversion_started_finished.groupby('course_title')
# Запишем названия курсов
groups_keys = conversion_started_finished_grouped.groups.keys()
```

```python
conversion_final = conversion_started_finished.copy()

for group in groups_keys:
    # Выберем одну группу
    current_group = conversion_started_finished_grouped.get_group(group).reset_index(drop=True)
    for i in range(1, len(current_group)): # i - номер модуля в данной группе
        finished = current_group.loc[i - 1, 'finished']
        started = current_group.loc[i, 'started']
        # Посчитаем конверсию для данного модуля, исключив деление на 0
        conversion = round((started / finished), 3) * 100 if finished != 0 else 0
        # Вписываем в датафрейм конверсию по изначальным индексам 
        index = conversion_started_finished_grouped.get_group(group).index[i]
        conversion_final.loc[index, 'conversion'] = conversion

# Выбросим модули, для которых конверсию посчитать невозможно
conversion_final.dropna(inplace=True)
conversion_final.head()
```

```python
fig, ax = plt.subplots(nrows=15, ncols=1, figsize=(15,50))
fig.suptitle('Конверсию перехода студентов из одного модуля в другой для каждого курса', y=1, fontsize=16)

for ax, course in zip(np.ravel(ax), groups_keys_fields):
        # Выберем группу по названию курса
        current_group = conversion_grouped.get_group(course[1])
        # Отрисовка графика, цвет выбираем в соответсвии со сферой курса
        ax.barh(current_group.module, current_group.conversion,\
                color=current_group.graph_color,\
                label=course[0])
        # Посчитаем для группы медианную конверсию
        conversion_median = current_group.conversion.median()
        # Добавим линию медианы
        ax.axvline(conversion_median, color='#2D2D2A', alpha=0.7, ls='--', lw=2,\
                   label=f'Медиана = {round(conversion_median, 1)}')
        # Оформление графика
        ax.set_title(course[1], fontsize=15)
        ax.set_xlabel('Конверсия, %')
        ax.legend(bbox_to_anchor=(1, 0.5), borderaxespad=3, framealpha=0.1)
        ax.set_frame_on(False)
        ax.invert_yaxis()

plt.tick_params(labelsize=9)
plt.tight_layout()
plt.show()
```
